# Metric #1: Cell Tower Locations in Allegheny County

Start off by reading the data from Pittsburgh neighborhoods

```python
import geopandas as gpd
import pandas as pd

neighborhood_boundaries = gpd.read_file('./Neighborhoods_.shp')

neighborhood_boundaries.head(4)
```

Here is the data for Pittsburgh neighborhoods, with each neighborhood name being found under the "hood" column and their borders under the "geometry" column.

Now we have to load in the data for the cell tower points.

```python
towers = gpd.read_file('./CellTowerPoints.shp') 

#Renaming columns to "longitude" and "latitude"
towers = towers.rename(columns={"Longitude_X": "Longitude", "Latitude_Y": "Latitude"})

towers.head(4)
```

Here is our the cell tower data, with the points under the "geometry" column using latitude and longitude

Cleaning up the data, so we only see the neighborhood, its coordinates, and its number of cell towers.

```python
import geopandas as gpd
import pandas as pd


# Create a new DataFrame for neighborhoods
neighborhoods = pd.DataFrame(columns=["Neighborhood", "Geometry", "Towers"])
neighborhoods["Neighborhood"] = neighborhood_boundaries["hood"]
neighborhoods["Geometry"] = neighborhood_boundaries["geometry"]
neighborhoods["Towers"] = 0

neighborhoods.head(3)
```

We are now going to iterate through the cell towers in order to find out which neighborhoods have the most cell towers.

```python
from shapely.geometry import Point

for i, tower in towers.iterrows():
    # Create a Point geometry for the tower
    tower_pt = tower["geometry"]
    # Iterate through each neighborhood boundary
    for j, neighborhood in neighborhoods.iterrows():
        # Check if the neighborhood contains the tower
        if neighborhood["Geometry"].contains(tower_pt):
            neighborhoods.at[j, 'Towers'] = neighborhood["Towers"] + 1
            # Break to save computation time
            break

neighborhoods.head(3)
```

Now, we are able to sort the data.

```python
neighborhoods_sorted = neighborhoods.sort_values("Towers", ascending=False)

neighborhoods_sorted.head()
```

So, as we can see, Hazelwood has the most cell towers, with a huge number of 4.

Now, we are going to plot the top 20 neighborhoods. There probably won't be much variety in this plot, though, so we are only plotting 20.



```python
plt = neighborhoods_sorted.iloc[:20].plot(kind='bar', x="Neighborhood", y="Towers", title="Cell Towers by Neighborhood")
```

Now we can plot the bottom 20. We had to use boolean indexing to exclude neighborhoods with 0 towers, since none existed in Allegheny County.

```python
# Create a new DataFrame with neighborhoods that have at least 1 tower
neighborhoods_nonzero = neighborhoods[neighborhoods["Towers"] > 0]

# Sort neighborhoods by the number of towers in ascending order
neighborhoods_asc = neighborhoods_nonzero.sort_values("Towers")

# Plot the bottom 20 neighborhoods by number of towers
plt2 = neighborhoods_asc.iloc[:20].plot(kind='bar', x="Neighborhood", y="Towers", title="Cell Towers by Neighborhood (Bottom 20)")
plt2.set_xlabel("Neighborhood")
plt2.set_ylabel("Number of Cell Towers")
plt2.tick_params(axis='x', labelrotation=90)

plt2.plot()
```

Now, we are going to add the cell tower data to the "neighborhood_boundaries" GeoDataFrame

```python
neighborhood_boundaries["Towers"] = neighborhoods["Towers"]

plt3 = neighborhood_boundaries.plot(column="Towers", legend=True)   
```

Finally, we can make a final ranking through the creation of a new dataframe.

```python
# Create a new dataframe to store the final ranks
neighborhoods_tower_ranked = pd.DataFrame()

# Add the neighborhood names to the new dataframe
neighborhoods_tower_ranked["Neighborhood"] = neighborhoods["Neighborhood"]

# Create ranks for each of the metrics
neighborhoods_tower_ranked["Towers_Rank"] = 0

# Sort neighborhoods by number of cell towers and assign ranks
neighborhoods_sorted = neighborhoods.sort_values("Towers", ascending=False)
rank = 1
for index, neighborhood in neighborhoods_sorted.iterrows():
    neighborhoods_tower_ranked.loc[neighborhoods_tower_ranked["Neighborhood"] == neighborhood["Neighborhood"], "Towers_Rank"] = rank
    rank += 1

# Display the top 10 neighborhoods by cell tower count with their corresponding ranks
neighborhoods_tower_ranked.sort_values("Towers_Rank").head(10)
```

# Metric 2: Number of Magisterial District Judge Offices per Neighborhood

First we have to load in the data from the pittsburgh Neighborhods dataset again.

```python
import geopandas as gpd
import pandas as pd

neighborhood_regions = gpd.read_file('./Neighborhoods_.shp')

neighborhood_regions.head(4)
```

Once again, each neighborhood's name is found under the "hood" column and their borders under the "geometry" column.

Now we can load in the data for landslides.

```python
import geopandas as gpd

offices = gpd.read_file('./MDJOffices.shp') 

#Renaming columns to "longitude" and "latitude"
offices = offices.rename(columns={"Longitude_X": "Longitude", "Latitude_Y": "Latitude"})

offices.head(4)
```

This data table tells us the name of the magisterial district, its judge, along with its location as longitude and latitude, which is found under "geometry"

Once again we have to clean up the data.

```python
import pandas as pd
import geopandas as gpd

# Renaming columns to "longitude" and "latitude"
offices = offices.rename(columns={"Longitude_X": "Longitude", "Latitude_Y": "Latitude"})

# Create a new DataFrame for neighborhoods
neighborhoods = pd.DataFrame(columns=["Neighborhood", "Geometry", "Offices"])
neighborhoods["Neighborhood"] = neighborhood_regions["hood"]
neighborhoods["Geometry"] = neighborhood_regions["geometry"]
neighborhoods["MDJ"] = 0

# Loop through each neighborhood and count the number of offices
for i, neighborhood in neighborhoods.iterrows():
    neighborhood_boundary = neighborhood["Geometry"]
    num_offices = len(offices[offices.within(neighborhood_boundary)])
    neighborhoods.at[i, 'Offices'] = num_offices

# Display the first 5 rows of the resulting DataFrame
neighborhoods.head(5)
```

Now, we are going to once again iterate through the magisterial district judge offices to find out which neighborhood has the most.

```python
from shapely.geometry import Point

neighborhoods["MDJ"] = 0  # add a new column 'Office' and set all values to 0

for i, office in offices.iterrows():
    # Create a Point geometry for the office
    office_pt = office["geometry"]
    # Iterate through each neighborhood boundary
    for j, neighborhood in neighborhoods.iterrows():
        # Check if the neighborhood contains the office
        if neighborhood["Geometry"].contains(office_pt):
            neighborhoods.at[j, 'MDJ'] += 1  # increment the count for the matching neighborhood
            # Break to save computation time
            break

neighborhoods.head(3)
```

Now, we can once again sort the data

```python
neighborhoods_sorted = neighborhoods.sort_values("Offices", ascending=False)
neighborhoods_sorted.head()
```

Now, we can plot the top 30 neighborhoods

```python
#import matplotlib
import matplotlib.pyplot as plt

# Sort neighborhoods by number of offices in descending order
neighborhoods_sorted = neighborhoods.sort_values("Offices", ascending=False)

# Plot the top 30 neighborhoods by number of offices
plt.bar(x=neighborhoods_sorted.iloc[:30]["Neighborhood"], height=neighborhoods_sorted.iloc[:30]["Offices"])
plt.xticks(rotation=90)
plt.title("Magisterial District Judges by Neighborhood (Top 30)")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Magisterial District Judges")
plt.show()
```

Since many neighborhoods don't have magisterial district judges offices, we don't need to plot the bottom 20 neighborhoods.

```python
neighborhood_regions["Offices"] = neighborhoods["Offices"]
plt3 = neighborhood_regions.plot(column="Offices", legend=True) 
```

Finally, we can do a final ranking of our top ten neighborhoods.

```python
# Create a new dataframe to store the final ranks
neighborhoods_ranked = pd.DataFrame()

# Add the neighborhood names to the new dataframe
neighborhoods_ranked["Neighborhood"] = neighborhoods["Neighborhood"]

# Create ranks for each of the metrics
neighborhoods_ranked["MDJ_Rank"] = 0

# Sort neighborhoods by number of magisterial district judges and assign ranks
neighborhoods_sorted = neighborhoods.sort_values("MDJ", ascending=False)
rank = 1
for index, neighborhood in neighborhoods_sorted.iterrows():
    neighborhoods_ranked.loc[neighborhoods_ranked["Neighborhood"] == neighborhood["Neighborhood"], "MDJ_Rank"] = rank
    rank += 1

# Display the top 10 neighborhoods by magisterial district judge count with their corresponding ranks
neighborhoods_ranked.sort_values("MDJ_Rank").head(10)
```

It seems we have a tie with our top 12 neighborhoods which all have one magisterial district judge office.
